```python
"""교통량 데이터"""
from tqdm import tqdm
API_KEY = '566645434c626c7539316c726d5454'
traffic = pd.DataFrame()

spotlist = ['B-12','C-18','C-20','D-44','D-45']
month = ["%.2d" % i for i in range(1,7)]
daylist = ["%.2d" % j for j in range(1,32)]
timelist = ["%.2d" % k for k in range(1,25)]

#url 입력
for spot in tqdm(spotlist) :
    for mon in tqdm(month) : 
        for day in daylist :
            for time in timelist :
                URL = 'http://openapi.seoul.go.kr:8088/566645434c626c7539316c726d5454/xml/VolInfo/1/999/{}/2023{}{}/{}/'.format(spot,mon,day,time)
                params ={'KEY' : '566645434c626c7539316c726d5454', 
                'TYPE' : 'xml', 
                'SERVICE' : 'VolInfo', 
                'START_INDEX' : 1, 
                'END_INDEX' : 9999}

                response = requests.get(URL, params=params)
                contents = response.text
                xml_obj = bs4.BeautifulSoup(contents,'lxml-xml')
                rows = xml_obj.findAll('row')

                # 각 행의 컬럼, 이름, 값을 가지는 리스트 만들기
                row_list = [] # 행값
                name_list = [] # 열이름값
                value_list = [] #데이터값

                # xml 안의 데이터 수집
                for i in range(0, len(rows)):
                    columns = rows[i].find_all()
                    #첫째 행 데이터 수집
                    for j in range(0,len(columns)):
                        if i ==0:
                            # 컬럼 이름 값 저장
                            name_list.append(columns[j].name)
                        # 컬럼의 각 데이터 값 저장
                        value_list.append(columns[j].text)
                    # 각 행의 value값 전체 저장
                    row_list.append(value_list)
                    # 데이터 리스트 값 초기화
                    value_list=[]

                dataframe = pd.DataFrame(row_list, columns=name_list)
                print(dataframe.tail(3))
                traffic = pd.concat([traffic , dataframe])
```
